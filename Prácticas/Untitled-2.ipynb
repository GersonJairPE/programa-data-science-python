```python
import pandas as pd
import numpy as np
```

```python
np.random.seed(123456)

df3 = pd.DataFrame(
    data=np.random.randint(0, 11, (10,5)),
    index=pd.date_range("20220101", periods=10), # Index
    columns=["a", "b", "c", "d", "e"]
    )
#df3
df3['ID'] = ["id" + str(i) for i in range(1,11)]  # [ > a√±adiendo una columna al DataFrame
df3
```

```python
df4 = df3.copy() 
df4["f"] = ["Femenino", "Masculino", "Masculino", "Femenino", "Masculino", "Femenino", "Masculino", "Femenino", "Masculino", "Femenino"]

df4
```

```python
df6 = df4.copy()
df6["g"] = ["R", "N", "R", "N", "R", "N", "N", "N", "N", "N"]  # a√±adimos una columna cualitativa

lc = ["f", "g"]  # lista de columnas para agrupamiento          
col_nums = ["a",	"b",	"e"]
df6.groupby(lc)[col_nums].max().round(2)
```

```python
df6.groupby("e").size()/10*100
```
